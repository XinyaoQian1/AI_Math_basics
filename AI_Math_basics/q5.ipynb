```python
#x
x = np.linspace(1,100,100)
print(x)
```

```python
#y
y = 5 * x + 10
print(y)
```

```python
plt.plot(x,y)
```

```python
y2 = y+ 50 * np.random.randn(100)
print(y2)
plt.plot(x,y2)
```

```python
plt.plot(x,y2,'.k')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#生成模型
model = LinearRegression()
model.fit(x,y2)
```

```python
x = x.reshape(100,1)
y2 = y2.reshape(100,1)
```

```python
model.fit(x,y2)
```

```python
y_predict = model.predict(x)
```

```python
plt.plot(x,y2,'.k')
plt.plot(x,y_predict,'b')
```

```python
#输入自变量10，让模型进行预测
y_10 = model.predict(np.array(10).reshape(1,1))
print('The predict value is',y_10)
```
